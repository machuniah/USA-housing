```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("usa_housing_kaggle.csv")  # Update with your file path

# Create a joint plot to visualize the relationship between CrimeRate and Price
plt.figure(figsize=(8, 6))
sns.jointplot(x="CrimeRate", y="Price", data=df, kind="reg", color="red")

# Set title
plt.suptitle("Crime Rate vs Price (with Regression Line)", y=1.02)

# Show plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("usa_housing_kaggle.csv")  # Update the file path if needed

# Create a scatter plot with regression line for YearBuilt vs Price
plt.figure(figsize=(8, 6))
sns.jointplot(x="YearBuilt", y="Price", data=df, kind="reg", color="blue")

# Set title
plt.suptitle("Year Built vs Price (with Regression Line)", y=1.02)

# Show plot
plt.show()
```
